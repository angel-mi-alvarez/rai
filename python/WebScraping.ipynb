## WebScraping with Python

```python
from bs4 import BeautifulSoup
import urllib.request
from IPython.display import HTML
import re
```

```python
r=urllib.request.urlopen('https://www.morningstar.es/es/funds/snapshot/p_snapshot.aspx?id=F00000SX7I')
soup = BeautifulSoup(r,"lxml")
type(soup)
```

```python
# Ver el código HTML
print(soup.prettify())
```

```python
# Ver el Texto
print(soup.get_text())
```

```python
for palabra in soup.findAll('span'):
    print(palabra,"\n")
```

```python
# Graba en un fichero el resultado que me interese
file =open("../out/parsed_data.txt","w")
for link in soup.findAll('a', attrs={'href': re.compile("^http")}):
    soup_link=str(link)
    print(soup_link)
    file.write(soup_link)
file.flush()
file.close()
```
